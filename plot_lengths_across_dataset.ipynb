# Plot lengths across datasets

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data loading (replace with your actual data)
df1 = pd.read_csv('Feb_1_2024_MS_Train_Val_Datasets/MEDIQA-CORR-2024-MS-TrainingData.csv')
df2 = pd.read_csv('Feb_1_2024_MS_Train_Val_Datasets/MEDIQA-CORR-2024-MS-ValidationSet-1-Full.csv')
df3 = pd.read_csv('Feb_5_2024_UW_Validation_Set_Updated/MEDIQA-CORR-2024-UW-ValidationSet-1-Full_Feb.csv')
# Calculate the length of the "Text" column for each dataset
lengths1 = df1['Text'].str.len()
lengths2 = df2['Text'].str.len()
lengths3 = df3['Text'].str.len()

# Plotting
plt.figure(figsize=(10, 6))  # Adjust the size of your plot

# Histogram for each dataset
plt.hist(lengths1, bins=50, alpha=0.5, label='MS Train Set', density=True)
plt.hist(lengths2, bins=50, alpha=0.5, label='MS Validation', density=True)
plt.hist(lengths3, bins=50, alpha=0.5, label='UW Validation', density=True)
mean_length1 = lengths1.mean()
mean_length2 = lengths2.mean()
mean_length3 = lengths3.mean()

plt.axvline(mean_length1, color='blue', linestyle='dashed', linewidth=1, label=f'Mean length MS train: {mean_length1:.2f}')
plt.axvline(mean_length2, color='red', linestyle='dashed', linewidth=1, label=f'Mean length MS Val: {mean_length2:.2f}')
plt.axvline(mean_length3, color='green', linestyle='dashed', linewidth=1, label=f'Mean length UW Val: {mean_length3:.2f}')

# Adding labels and title
plt.xlabel('Length of Clinical Text')
plt.ylabel('Frequency')
plt.legend()

# Display the plot
plt.tight_layout()
plt.grid()
plt.savefig('lengths.png')
plt.show()
```
